# Select your features (columns)

```python
selected_features = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_period_err1', 'koi_period_err2',
       'koi_time0bk', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact',
       'koi_impact_err1', 'koi_impact_err2', 'koi_duration',
       'koi_duration_err1', 'koi_duration_err2', 'koi_depth', 'koi_depth_err1',
       'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2',
       'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2',
       'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_steff_err1',
       'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2',
       'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec',
       'koi_kepmag']]
```

```python
X = selected_features
y = df['koi_disposition'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Test-Train Split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
# from tensorflow.keras.utils import to_categorical
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# find feature importance using random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=500)
rf = rf.fit(X,y)
```

```python
importances = rf.feature_importances_
importances
```

```python
sorted(zip(rf.feature_importances_, X), reverse=True)
```

```python
# drop the features that are not important

X = df[['koi_fpflag_co','koi_fpflag_nt','koi_fpflag_ss','koi_model_snr', 'koi_prad',
       'koi_fpflag_ec','koi_duration_err1','koi_prad_err2','koi_duration_err2',
       'koi_steff_err1','koi_prad_err1','koi_steff_err2','koi_time0bk_err1',
       'koi_duration', 'koi_time0bk_err2','koi_period']]
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42)
```

```python
# Scale your data using whatever scaler
# X_scaler = MinMaxScaler().fit(X_train)
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# y_scaler = StandardScaler().fit(y_train)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)
```

# Train various  models



## Model 1 Logistic Regression Scaled

```python
# Instantiate model 1 (e.g., logistic model)
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=500)

# Fit model using GridSearchCV for hyperparameter turning or not (logistic I wouldn't use gridsearch)
model.fit(X_train_scaled, y_train)
```

```python
# Look at model evalauation metrics

# Includes score, but also confusion matrix and classification report
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")

```

```python
# predictions = model.predict(X_test_scaled)
# predictions.reshape(-1, 1)
# X_scaler.inverse_transform(predictions)
predictions = model.predict(X_test_scaled)
predictions
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(predictions, y_test, labels=['FALSE POSITIVE', 'CANDIDATE', 'CONFIRMED'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

## Model 3

```python
# Instantiate model 2 (e.g., random forest model)
rf_model = RandomForestClassifier()

# Fit model using GridSearchCV for hyperparameter turning or not (logistic I wouldn't use gridsearch)
rf_model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {rf_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf_model.score(X_test_scaled, y_test)}")
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
```

```python
rf_model.get_params()
```

```python
# param_grid = {'C': [1, 5, 10, 50],
#               'gamma': [0.0001, 0.0005, 0.001, 0.005]}
param_grid = {'min_samples_split': [0, 1, 2, 3, 4, 5],
              'min_samples_leaf': [0, 1, 2, 3, 4, 5],
              'n_estimators': [25, 50, 100, 200, 250, 300]
             }
grid = GridSearchCV(rf_model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
# param_grid = {'min_samples_split': [3],
#               'min_samples_leaf': [2],
#               'n_estimators': [200]
#              }
param_grid = {'min_samples_split': [3],
              'min_samples_leaf': [3],
              'n_estimators': [300]
             }
grid = GridSearchCV(rf_model, param_grid, verbose=3)
grid.fit(X_train_scaled, y_train)
```

```python
grid_preds = grid.predict(X_test_scaled)
grid_preds
```

```python
confusion_matrix(grid_preds, y_test, labels=['FALSE POSITIVE', 'CANDIDATE', 'CONFIRMED'])
```

```python
print(classification_report(y_test, grid_preds))

```
