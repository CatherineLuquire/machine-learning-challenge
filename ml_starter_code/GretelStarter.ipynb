```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42)
```

```python
# Scale your data using whatever scaler
# X_scaler = MinMaxScaler().fit(X_train)
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train various  models



## Model 1

```python
# Instantiate model 1 (e.g., logistic model)
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=500)

# Fit model using GridSearchCV for hyperparameter turning or not (logistic I wouldn't use gridsearch)
model.fit(X_train_scaled, y_train)
```

```python
# Look at model evalauation metrics

# Includes score, but also confusion matrix and classification report
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")

```

```python
predictions = model.predict(X_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(predictions, y_test, labels=['FALSE POSITIVE', 'CANDIDATE', 'CONFIRMED'])
```

```python
preds=pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
preds
```

```python
preds['Prediction'].unique()
```

```python
preds['Actual'].unique()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["FALSE POSITIVE", "CANDIDATE", "CONFIRMED"]))
```

## Model 2

```python
# Instantiate model 2 (e.g., random forest model)
rf_model = RandomForestClassifier()

# Fit model using GridSearchCV for hyperparameter turning or not (logistic I wouldn't use gridsearch)
rf_model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {rf_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf_model.score(X_test_scaled, y_test)}")
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
```

```python
rf_model.get_params()
```

```python
# param_grid = {'C': [1, 5, 10, 50],
#               'gamma': [0.0001, 0.0005, 0.001, 0.005]}
param_grid = {'min_samples_split': [1, 2, 3],
              'min_samples_leaf': [0, 1, 2],
              'n_estimators': [100, 200, 300]
             }
grid = GridSearchCV(rf_model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
grid_preds = grid.predict(X_test_scaled)
grid_preds
```

```python
confusion_matrix(grid_preds, y_test, labels=['FALSE POSITIVE', 'CANDIDATE', 'CONFIRMED'])
```

```python
# Look at model evalauation metrics
rf_preds = rf_model.predict(X_test_scaled)
rf_preds
# Includes score, but also confusion matrix and classification report
```

```python
confusion_matrix(rf_preds, y_test, labels=['FALSE POSITIVE', 'CANDIDATE', 'CONFIRMED'])
```

```python
print(classification_report(y_test, rf_preds,
                            target_names=["FALSE POSITIVE", "CANDIDATE", "CONFIRMED"]))
```
